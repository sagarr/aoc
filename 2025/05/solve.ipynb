```python
with open('input.txt') as f:
    rng, ids = f.read().split("\n\n")
    
    rng = list(map(lambda x: (int(x[0]), int(x[1])), map(lambda x: x.split('-'), rng.strip().split())))
    ids = list(map(int, ids.strip().split()))
```

```python
# Part 1
```

```python
master_range = []

rng.sort()

ps, pe = rng[0]

for s, e in rng[1:]:
    if pe >= s:
        # overlap, extend end
        pe = max(pe, e)
    else:
        # separate segment, add
        master_range.append((ps, pe))
        ps, pe = s, e

master_range.append((ps, pe))
```

```python
count = 0

for id in ids:
    for s, e in master_range:
        if id < s:
            break
        if s <= id <= e:
            count+=1
            break

count
```

```python
# Part 2
```

```python
sum(e-s+1 for s, e, in master_range)
```
