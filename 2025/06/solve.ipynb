```python
problems = []
with open('input.txt') as f:
    for p in f.read().strip().splitlines():
        problems.append(p.strip().split())

R, C = len(problems), len(problems[0])
```

```python
# Part 1
```

```python
import math

res = 0

for c in range(C):
    nums = []
    for r in range(R-1):
        nums.append(int(problems[r][c]))
    
    if problems[-1][c] == '*':
        res += math.prod(nums)
    else:
        res += sum(nums)

res
```

```python
problems = []

with open('input.txt') as f:
    for p in f.read().strip().splitlines():
        problems.append(p)

R, C = len(problems), len(problems[0])
```

```python
human_math_list = []
num_list = []
for j in range(C - 1, -1, -1):
    if all(problems[r][j].isspace() for r in range(R)):
        human_math_list.append(num_list)
        num_list = []
        continue
    num = ''
    for i in range(R):
        try:
            if problems[i][j].isnumeric():
                num += problems[i][j]
            if problems[i][j] in '+*':
                num_list.insert(0, problems[i][j])
        except:
            print(f"culprit: {j=}")
    num_list.append(num)
    if j == 0:
        human_math_list.append(num_list)
```

```python
res = 0
for problem in human_math_list:
    if problem[0] == '+':
        res += sum(map(int, problem[1:]))
    else:
        res += math.prod(map(int, problem[1:]))

res
```
