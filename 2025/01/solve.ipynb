```python
with open('input.txt') as f:
    lines = f.read().splitlines()
```

# Part 1

```python
point = 50
password = 0

for line in lines:
    dir, p = line[:1], int(line[1:])
    
    if dir == 'R':
        point+=p
    else:
        point-=p

    point%=100
    
    if point == 0:
        password+=1

password
```

# Part 2

```python
point = 50
password = 0

for line in lines:
    dir, p = line[:1], int(line[1:])
    
    if dir == 'R':
        point+=p
    else:
        prev_point = point
        point-=p

    if point > 99:
        password += point//100
    elif point < 1:
        password += ((prev_point-1)//100) - ((point-1)//100)

    point%=100
    
password
```
