```python
grid = []

with open('input.txt') as f:
    for line in f.readlines():
        grid.append(line.strip())
```

```python
# Part 1
```

```python
from collections import defaultdict

beams = defaultdict(int)
beams[grid[0].index("S")] = 1
splits = 0
for r in range(1, len(grid)):
    for c in list(beams.keys()):
        if grid[r][c] == "^":
            beams[c-1] += beams[c]
            beams[c+1] += beams[c]
            del beams[c]
            splits+=1

splits
```
