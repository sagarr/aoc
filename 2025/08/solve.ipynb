```python
coords = []

with open('input.txt') as f:
    for c in f.readlines():
        coords.append(list(map(int, c.strip().split(','))))
```

```python
# Part 1
```

```python
class UnionFind:

    def __init__(self, n):
        self.parents = list(range(n))
        self.rank = {i: 1 for i in range(n)}
        self.circuits = [{i} for i in range(n)]

    def find(self, node):
        while self.parents[node] != node:
            node = self.parents[node]
        return node

    def union(self, u, v):
        up, vp = self.find(u), self.find(v)
        if up != vp:
            # merge
            if self.rank[up] > self.rank[vp]:
                self.rank[up]+=self.rank[vp]
                self.parents[vp] = up
                del self.rank[vp]
                self.circuits[up] |= self.circuits[vp]
            else:
                self.rank[vp]+=self.rank[up]
                self.parents[up] = vp
                del self.rank[up]

    def size(self):
        return list(self.rank.values())


def dist(c1, c2):
    dx = (coords[c1][0] - coords[c2][0]) ** 2
    dy = (coords[c1][1] - coords[c2][1]) ** 2
    dz = (coords[c1][2] - coords[c2][2]) ** 2
    return dx + dy + dz

n = len(coords)
distances = []

for c1 in range(n):
    for c2 in range(c1+1, n):
        distances.append([dist(c1, c2), (c1, c2)])

distances.sort(key=lambda x: x[0])

uf = UnionFind(n)

for _d, (c1, c2) in distances[:1000]:
    uf.union(c1, c2)

s = sorted(uf.size())
s.pop() * s.pop() * s.pop()
```
