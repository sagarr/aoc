```python
with open('input.txt') as f:
    ranges = f.read().strip().split(",")

ranges
```

# Part 1

```python
invalid_ids_total = 0

for rng in ranges:
    fr, to = rng.split("-")
    for i in range(int(fr), int(to)+1):
        s = str(i)
        if len(s) % 2 == 0:
            h = len(s)//2
            if s[:h] == s[h:]:
                invalid_ids_total += i
                
invalid_ids_total
```

# Part 2

```python
invalid_ids_total = 0

for rng in ranges:
    fr, to = rng.split("-")
    for i in range(int(fr), int(to)+1):
        s = str(i)

        half = len(s) // 2

        for j in range(1, half+1):
            num = len(s) // j
            if s[:j]*num == s:
                invalid_ids_total += i
                break

invalid_ids_total
```
