# Part 1

```python
grid = []

with open('input.txt') as f:
    for row in f.readlines():
      grid.append(list(row.strip()))

R, C = len(grid), len(grid[0])
```

```python
def floodfill(r, c, plant, found):
    if r < 0 or r >= R or c < 0 or c >= C:
        return 0, 0

    if grid[r][c] != plant:
        return 0, 0

    if (r, c) in found:
        return 0, 0

    found.add((r, c))

    perimeter = 0
    for dr, dc in [(1, 0), (-1, 0), (0, 1), (0, -1)]:
        nr, nc = r + dr, c + dc
        if (nr < 0 or nr >= R or nc < 0 or nc >= C or grid[nr][nc] != plant):
            perimeter += 1

    area = 1
    for dr, dc in [(1, 0), (-1, 0), (0, 1), (0, -1)]:
        a, p = floodfill(r + dr, c + dc, plant, found)
        area += a
        perimeter += p

    return area, perimeter

visited = set()
total_price = 0

for r in range(R):
    for c in range(C):
      if (r, c) not in visited:
        plant = grid[r][c]
        found = set()
        area, perimeter = floodfill(r, c, plant, found)
        total_price += area * perimeter
        visited.update(found)

total_price
```

```python
def floodfill(r, c, plant, found):
    if r < 0 or r >= R or c < 0 or c >= C:
        return 0, 0

    if grid[r][c] != plant:
        return 0, 0

    if (r, c) in found:
        return 0, 0

    found.add((r, c))

    perimeter = 0
    for dr, dc in [(1, 0), (-1, 0), (0, 1), (0, -1)]:
        nr, nc = r + dr, c + dc
        if nr < 0 or nr >= R or nc < 0 or nc >= C or grid[nr][nc] != plant:
            perimeter += 1

    area = 1
    for dr, dc in [(1, 0), (-1, 0), (0, 1), (0, -1)]:
        a, p = floodfill(r + dr, c + dc, plant, found)
        area += a
        perimeter += p

    return area, perimeter


def count_corners(found):
    left = set()
    right = set()
    up = set()
    down = set()
    for r, c in found:
        if (r - 1, c) not in found:
            up.add((r, c))

        if (r + 1, c) not in found:
            down.add((r, c))

        if (r, c + 1) not in found:
            right.add((r, c))

        if (r, c - 1) not in found:
            left.add((r, c))

    corners = 0
    for r, c in up:
        if (r, c) in left:
            corners += 1
        if (r, c) in right:
            corners += 1
        if (r - 1, c - 1) in right and (r, c) not in left:
            corners += 1
        if (r - 1, c + 1) in left and (r, c) not in right:
            corners += 1

    for r, c in down:
        if (r, c) in left:
            corners += 1
        if (r, c) in right:
            corners += 1
        if (r + 1, c - 1) in right and (r, c) not in left:
            corners += 1
        if (r + 1, c + 1) in left and (r, c) not in right:
            corners += 1

    return corners

visited = set()
total_price = 0

visited = set()
total_price = 0

for r in range(R):
    for c in range(C):
      if (r, c) not in visited:
        plant = grid[r][c]
        found = set()
        area, perimeter = floodfill(r, c, plant, found)
        corners = count_corners(found)
        total_price += corners * area
        visited.update(found)

total_price
```
