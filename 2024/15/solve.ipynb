# Part 1

```python
grid = []
with open('input.txt') as f:
  positions, moves = f.read().split("\n\n")
  for line in positions.split("\n"):
    grid.append(list(line))

  moves = moves.split("\n")
  moves = "".join(moves)
  moves = list(moves.strip())

print(len(moves))
```

```python
R, C = len(grid), len(grid[0])

robot_pos = None

for r in range(R):
    for c in range(C):
        if grid[r][c] == '@':
            robot_pos = (r, c)
            break

def print_grid(grid):
    for r in range(len(grid)):
        for c in range(len(grid[0])):
            print(grid[r][c], end='')
        print()

print_grid(grid)
```

```python
r, c = robot_pos

def move_box(move, r, c):
  if move == '>':
    while c < C-1 and grid[r][c+1] != '#' and grid[r][c+1] != '.':
      c += 1
    return c+1

  if move == '<':
    while c > 0 and grid[r][c-1] != '#' and grid[r][c-1] != '.':
      c -= 1
    return c-1

  if move == '^':
    while r > 0 and grid[r-1][c] != '#' and grid[r-1][c] != '.':
      r -= 1
    return r-1

  if move == 'v':
    while r < R-1 and grid[r+1][c] != '#' and grid[r+1][c] != '.':
      r += 1
    return r+1

for move in moves:
  if move == '>':
    if grid[r][c+1] == '.':
      grid[r][c] = '.'
      c += 1
      grid[r][c] = '@'
    elif grid[r][c+1] == 'O':
      nc = move_box(move, r, c+1)
      if nc >= C-1 or grid[r][nc] == '#':
        continue
      grid[r][c] = '.'
      c += 1
      grid[r][c] = '@'
      grid[r][nc] = 'O'
  elif move == '<':
    if grid[r][c-1] == '.':
      grid[r][c] = '.'
      c -= 1
      grid[r][c] = '@'
    elif grid[r][c-1] == 'O':
      nc = move_box(move, r, c-1)
      if nc <= 0 or grid[r][nc] == '#':
        continue
      grid[r][c] = '.'
      c -= 1
      grid[r][c] = '@'
      grid[r][nc] = 'O'
  elif move == '^':
    if grid[r-1][c] == '.':
      grid[r][c] = '.'
      r -= 1
      grid[r][c] = '@'
    elif grid[r-1][c] == 'O':
      nr = move_box(move, r-1, c)
      if nr <= 0 or grid[nr][c] == '#':
        continue
      grid[r][c] = '.'
      r -= 1
      grid[r][c] = '@'
      grid[nr][c] = 'O'
  elif move == 'v':
    if grid[r+1][c] == '.':
      grid[r][c] = '.'
      r += 1
      grid[r][c] = '@'
    elif grid[r+1][c] == 'O':
      nr = move_box(move, r+1, c)
      if nr >= R-1 or grid[nr][c] == '#':
        continue
      grid[r][c] = '.'
      r += 1
      grid[r][c] = '@'
      grid[nr][c] = 'O'
print_grid(grid)
```

```python
# gps
gps_sum = 0
for r in range(R):
  for c in range(C):
    if grid[r][c] == 'O':
      gps_sum += (r*100)+c

print(gps_sum)
```

```python
# expand
grid = []

def expand(ch):
  if ch == "#":
    return "##"
  if ch == "O":
    return "[]"
  if ch == ".":
    return ".."
  if ch == "@":
    return "@."

with open('input.txt') as f:
  positions, moves = f.read().split("\n\n")
  for line in positions.split("\n"):
    grid.append([expand(ch) for ch in line])
  moves = moves.split("\n")
  moves = "".join(moves)
  moves = list(moves.strip())

print_grid(grid)
print(moves)

R, C = len(grid), len(grid[0])
print(R, C)
```

```python

robot_pos = None

for r in range(R):
    for c in range(C):
        if grid[r][c] == '@.':
            robot_pos = (r, c)
            break

print(robot_pos)
```

```python
r, c = robot_pos

def move_box(move, r, c):
  if move == '>':
    while c < C-1 and grid[r][c+1] != '##' and grid[r][c+1] != '..':
      c += 1
    return c+1

  if move == '<':
    while c > 0 and grid[r][c-1] != '##' and grid[r][c-1] != '..':
      c -= 1
    return c-1

  if move == '^':
    while r > 0 and grid[r-1][c] != '##' and grid[r-1][c] != '..':
      r -= 1
    return r-1

  if move == 'v':
    while r < R-1 and grid[r+1][c] != '##' and grid[r+1][c] != '..':
      r += 1
    return r+1

for move in moves[:1]:
  if move == '>':
    if grid[r][c+1] == '..':
      grid[r][c] = '..'
      c += 1
      grid[r][c] = '@.'
    elif grid[r][c+1] == '[]':
      nc = move_box(move, r, c+1)
      if nc >= C-1 or grid[r][nc] == '##':
        continue
      grid[r][c] = '..'
      c += 1
      grid[r][c] = '@.'
      grid[r][nc] = '[]'
  elif move == '<':
    if grid[r][c-1] == '..':
      grid[r][c] = '..'
      c -= 1
      grid[r][c] = '@.'
    elif grid[r][c-1] == '[]':
      nc = move_box(move, r, c-1)
      if nc <= 0 or grid[r][nc] == '##':
        continue
      grid[r][c] = '.'
      c -= 1
      grid[r][c] = '@.'
      grid[r][nc] = '[]'
  elif move == '^':
    if grid[r-1][c] == '..':
      grid[r][c] = '..'
      r -= 1
      grid[r][c] = '@.'
    elif grid[r-1][c] == '[]':
      nr = move_box(move, r-1, c)
      if nr <= 0 or grid[nr][c] == '##':
        continue
      grid[r][c] = '..'
      r -= 1
      grid[r][c] = '@.'
      grid[nr][c] = '[]'
  elif move == 'v':
    if grid[r+1][c] == '..':
      grid[r][c] = '..'
      r += 1
      grid[r][c] = '@.'
    elif grid[r+1][c] == '[]':
      nr = move_box(move, r+1, c)
      if nr >= R-1 or grid[nr][c] == '##':
        continue
      grid[r][c] = '..'

      r += 1
      grid[r][c] = '@.'
      grid[nr][c] = '[]'
  print(move, r, c)
print_grid(grid)
```

```python
warehouse, moves = open('input.txt').read().split('\n\n')
warehouse, robot = [[('[' if i % 2 == 0 else ']') if line[i // 2] == 'O' else line[i // 2] for i in range(len(line) * 2)] for line in warehouse.split()], warehouse.index('@')
x, y, directions = robot % (len(warehouse[0]) // 2), robot // (len(warehouse[0]) // 2), {'>': 1, '<': -1, '^': -1, 'v': 1}
warehouse[y][x], warehouse[y][x + 1] = '.', '.'
for move in ''.join(moves.split()):
    if move in ['>', '<']:
        x1 = x + (directions[move])
        while warehouse[y][x1] in ['[', ']']: x1 += directions[move]
        if warehouse[y][x1] == '.':
            for x2 in range(x1, x, -directions[move]): warehouse[y][x2] = warehouse[y][x2 - directions[move]]
            x += directions[move]
    else:
        boxes = [{(x, y)}]
        while boxes[-1]:
            boxes.append(set())
            for box in boxes[-2]:
                if warehouse[box[1] + directions[move]][box[0]] == '#': break
                if warehouse[box[1] + directions[move]][box[0]] == '[': boxes[-1] |= {(box[0], box[1] + directions[move]), (box[0] + 1, box[1] + directions[move])}
                elif warehouse[box[1] + directions[move]][box[0]] == ']': boxes[-1] |= {(box[0], box[1] + directions[move]), (box[0] - 1, box[1] + directions[move])}
            else: continue
            break
        else:
            for row in list(reversed(boxes)):
                for box in row: warehouse[box[1] + directions[move]][box[0]], warehouse[box[1]][box[0]] = warehouse[box[1]][box[0]], '.'
            y += directions[move]
print(sum([100 * i + j for i, line in enumerate(warehouse) for j, c in enumerate(line) if c == '[']))
```
